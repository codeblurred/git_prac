```python
x=6
x %= 3
print(x)

a=6
a //= 3
print(a)

b=6
b **= 3
print(b)

c=6
c &= 3
print(c)

d=6
d |= 3
print(d)


e=6
e ^= 3
print(e)


f=6
f >>= 3
print(f)

g=6
g <<= 3
print(g)
```

```python
# comparison operators
a = 3
b = 5

print(a == b)
print(a != b)
print( a > b)
print( a >= b)
print(a < b)
print(a <= b)
```

```python
# identity operators
a = ["cadbury","paracetamol"]
c = ["cadbury","paracetamol"]

b = a

# print(a,b,c)

print(a is c)

print(a is b)

print(a is not b)

print(a == b)
```

```python
# membership operators
a = ["cadbury","paracetamol"]
m = a
print("suman" in a)
print("suman" not in m )
print("cadbury" in a)
```

```python
# logical operators 

x = 3

print(x > 1 and x < 6) 
print(x > 1 or x < 2)

print(not(x > 1 and x < 6))
print(not(x > 1 or x < 2))
```

```python
print(x > 1 & x < 6) 
print(x > 1 | x < 2)

# ~(not) << (zero fill left shift ) >> ^(XOR)
```
