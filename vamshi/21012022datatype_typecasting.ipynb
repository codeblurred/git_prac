```python
# text type : str 


# numeric : int float complex

# sequence : list , tuple , range

# mapping : dict

# set types: set, frozenset

# boolean : bool 
```

```python
# text type : str

x = "tukzer"
print(type(x))
```

```python
# numeric : int float complex
# int
w = 1
print(type(w))

# float
x = 1.0
print(type(x))

# complex
y = 1j
print(type(y))



```

```python
# sequence : list , tuple , range

a = ["apple","banana","cherry"]
print(type(a))

b = ("apple","banana","cherry")
print(type(b))

c = range(7)
print(type(c))
```

```python
# mapping : dict

map1 = {"name": "suman", "age": "30"}

print(type(map1))
```

```python
# set types: set, frozenset
set1 = {"apple","banana","cherry"}
print(type(set1))

set_frozen = frozenset({"apple","banana","cherry"})
print(type(set_frozen))
```

```python
# boolean : bool 
c = True
print(type(c))
```

```python
# Typecasting 

x = int(1)

print(type(x))

y = int(1.2)
print(type(y))
print(y)

z = float(x)
print(type(z))
print(z)


d = str(x)
print(type(d))
print(d)
```
