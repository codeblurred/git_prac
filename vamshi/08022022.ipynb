```python


def suman(n):
    return abs(n - 50)

new_mynum = [120, 136, 67, 87, 23, 25, 12, 7, 65]

new_mynum.sort(key = suman)

print(new_mynum)
```

```python
mylist = ['Banana', 'Grass', 'Suman', 'Zebra', 'buffalo', 'cow', 'eat', 'eats', 'suman', 'uS', 'us']

mylist.sort(key = str.lower)

#  1st priority capital letters 
# we can tweak through the key built in function

print(mylist)

```

```python
mylist = ['Banana', 'Grass', 'Suman', 'Zebra', 'buffalo', 'cow', 'eat', 'eats', 'suman', 'uS', 'us']

len(mylist)
range(len(mylist))
for item in range(len(mylist)):
    mylist[item] = mylist[item].lower()    
print(mylist)

mylist.sort()

print("the sorted list is :", mylist)
```
