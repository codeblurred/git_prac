```python
list1=["a","test1","2","5hsfdf","hello","apple","banana","jackfruit","elephant","horse"]

########### my goal is to print all the elements that had letter "e" in it 

new_list = []

for item in list1:  # part1
    if "e" in item: # part2
        new_list.append(item)
print(new_list)


# ################# 
# new_list1 = [item <<for item in list1>part 1>  <<if "e"  in item>part2>] 
new_list1 = [item for item in list1 if "e"  in item] 
print(new_list1) 

# syntax [expression=item for item in iteration=list1 if condition == true ]
```
